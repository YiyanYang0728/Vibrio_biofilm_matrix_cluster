```r
library(ggplot2)
library(ggpubr)
library(tidyverse)
```

```r
dt0 = read.csv("/data/yangy34/projects/vibrio_paper/prophage_VibrioBMC/BMC_type_vs2_prophage_ct.tsv", sep="\t")
head(dt0)
dt <- dt0 %>% mutate(new_type=ifelse(bmc_type == "none", "neg", "pos"))
my_comparisons <- list( c("pos", "neg"))
q <- ggboxplot(dt, x="new_type", y="prophage_ct", color="new_type", palette = "jco", add = "jitter") +
stat_compare_means(comparisons = my_comparisons)+ # Add pairwise comparisons p-value
  stat_compare_means(label.y = 45)
q
```

```r
dt %>% group_by(new_type) %>% summarize(Mean = mean(prophage_ct))
```

```r
dt1 <- dt0
my_comparisons <- list( c("complete", "partial"), c("partial", "none"), c("none", "complete") )
p <- ggboxplot(dt1, x="bmc_type", y="prophage_ct", color="bmc_type", palette = "jco", add = "jitter") +
stat_compare_means(comparisons = my_comparisons)+ # Add pairwise comparisons p-value
  stat_compare_means(label.y = 55)
p
```

```r
dt1 %>% group_by(bmc_type) %>% summarize(Mean = mean(prophage_ct))
```
